```python
#Fazer um jogo da forca criando uma lista choice(['leonardo,'barnabé,diogo])
#find & for
#transformar a palavra em Lista for String.append

import random

# Lista de palavras possíveis
frutas = ["banana", "morango", "abacaxi", "caqui","manga","laranja","maça","pêssego","romã","dragon fruit","ameixa","mexerica",
            "jaca","amora","damasco","abacaxi"]

# Escolhe uma palavra aleatoriamente
fruta = random.choice(frutas).upper()

# Inicializa a lista de letras acertadas com underscores
letras_acertadas = ['_' for letra in fruta]

# Número de tentativas
tentativas = 7

# Loop principal do jogo
while True:
    # Exibe a palavra parcialmente
    print(" ".join(letras_acertadas))

    # Pede ao usuário para digitar uma letra
    letra = input("Digite uma letra que corresponde o nome de uma fruta: ").upper()

    # Verifica se a letra está na palavra
    if letra in fruta:
        # Atualiza a lista de letras acertadas
        for i, letra_fruta in enumerate(fruta):
            if letra == letra_fruta:
                letras_acertadas[i] = letra
    else:
        # Reduz o número de tentativas
        tentativas -= 1
        print(f"Letra errada! Você tem {tentativas} tentativas restantes.")

    # Verifica se o jogador ganhou ou perdeu
    if "_" not in letras_acertadas:
        print(f"Parabéns! Você Acertou \o/ O nome da fruta é: {fruta}")
        break
    elif tentativas == 0:
        print(f"Você perdeu! T-T O nome correto da Fruta era: {fruta}")
        break
```

```python
import random

# Lista de palavras possíveis com nomes de animais
def escolher_animal():
    animais = ["gato", "cachorro", "cavalo", "porco", "zebra", "burro", "elefante",
               "camelo", "jirafa", "ovelha", "morcego", "cobra", "jacare", "golfinho",
               "foca", "rato"]
    return random.choice(animais).upper()  # Escolhe uma palavra aleatoriamente

# Inicializa a lista de letras acertadas com underscores
def exibir_animal(letras_acertadas):
    return " ".join(letras_acertadas)

# Função principal do jogo
def main():
    animal = escolher_animal()
    letras_acertadas = ['_' for _ in animal]
    tentativas = 7

    # Loop principal do jogo
    while True:
        print(exibir_animal(letras_acertadas))

        # Pede ao usuário para digitar uma letra
        letra = input("Digite uma letra que corresponde ao nome de um animal: ").upper()
        if len(letra) != 1 or not letra.isalpha():
            print("Por favor, insira apenas uma letra.")
            continue

        # Verifica se a letra está na palavra
        if letra in animal:
            for i, letra_palavra in enumerate(animal):
                if letra == letra_palavra:
                    letras_acertadas[i] = letra
        else:
            # Reduz o número de tentativas
            tentativas -= 1
            print(f"Letra errada! Você tem {tentativas} tentativas restantes.")

        # Verifica se o jogador ganhou ou perdeu
        if "_" not in letras_acertadas:
            print(f"Parabéns! Você acertou \o/ O nome do animal é: {animal}")
            break
        elif tentativas == 0:
            print(f"Você perdeu! T-T O nome correto do animal era: {animal}")
            break

if __name__ == "__main__":
    main()
```
